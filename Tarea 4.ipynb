**1) ¿Cual es el nivel educacional mas frecuente en la muestra y que porcentaje representa?**

- Variables involucradas: e6a (cualitativa)

```python
e6a_etiquetas = {1: 'Nunca asistió', 2: 'Sala cuna', 3: 'Jardín Infantil ', 4: 'Prekinder/Kinder', 5: 'Educación Especial', 6: 'Primaria o Preparatoria', 7: 'Educación Básica', 8: 'Humanidades', 9: 'Educación Media Científico Humanista', 10: 'Técnica, Comercial, Industrial o Normalista', 11: 'Educación Media Técnica Profesional', 12: 'Técnico Nivel Superior Incompleto (Carreras 1 a 3 años)', 13: "Técnico Nivel Superior Completo", 14: "Profesional Incompleto (Carreras 4 o más años)",15: "Profesional Completo (Carreras 4 o más años)", 16: "Postgrado Incompleto", 17: "Postgrado Completo", 99: "No sabe/no responde"}
datos['e6a_etiquetas'] = datos['e6a'].map(e6a_etiquetas)
print('*' * 20, "Tabla de frecuencia para la variable nivel educacional ", '*' * 20)
print(datos['e6a_etiquetas'].value_counts())
print('*' * 20, "Tabla de porcentajes para la variable nivel educacional", '*' * 20)
print(round(datos['e6a_etiquetas'].value_counts()/len(datos)*100,2))
```

```python
plt.figure(figsize=(4, 4))
ax = sns.countplot(x='e6a_etiquetas', data=datos)
plt.title('Gráfico de barras para la variable el nivel educacional maximo alcanzado', y=1)
plt.ylabel('Conteo')

ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

**Respuesta:**

- El nivel educacional mas frecuente en la muestra es Educación Media Científico Humanista. 
- El 26,14% de las personas encuestadas definio su nivel educacional más alto alcanzado o nivel educacional actual como Educación Media Científico Humanista. 
- Con respecto al grafico de barras, se confirma que el mayor nivel educacional fue Educación Media Científico Humanista y el menor nivel educacional es postgrado incompleto.

**2) ¿En que zonas se localiza la mayor cantidad de personas de sexo femenino?**

- Variables involucradas: sexo, urbano (cualitativo nominal-cualitativo nominal)

```python
sexo_etiquetas = {1: 'hombre', 2: 'mujer'} 
zona_etiquetas = {1: 'Urbano',2:'Rural'}
print('*' * 10, "Tabla de frecuencia de las variables Sexo vs Zona", '*' * 10)
tabla1 = pd.crosstab(datos['sexo'].replace(sexo_etiquetas), datos['zona'].replace(zona_etiquetas), margins=True, margins_name='Suma')
print(tabla1)

print('*' * 10, "Tabla de porcentajes de las variables Sexo vs Zona", '*' * 10)
tabla1=round(pd.crosstab(datos['sexo'].replace(sexo_etiquetas), datos['zona'].replace(zona_etiquetas),margins=True,
                         margins_name='Total porcentaje',normalize='index')*100,2)
print(tabla1)
```

```python
sexo_etiquetas = {1: 'hombre', 2: 'mujer'} 
zona_etiquetas = {1: 'Urbano',2:'Rural'}
plt.figure(figsize=(4,4))
count_plot = sns.countplot(x = datos['sexo'].replace(sexo_etiquetas), hue = datos['zona'].replace(zona_etiquetas))
plt.title('Gráfico de barras de localizacion de personas por genero ',y=1.01)
plt.ylabel('Conteo')
plt.xlabel('Sexo')
plt.show()
```

**Respuesta:**

- La mayor concentraccion de personas de sexo femenino, se localiza en zonas Urbanas.
- Del total de personas de sexo femenino estudiadas el 82,21% se localizan en zonas urbanas.
- Con respecto al grafico, se confirma que en ambos sexos predomina la zona urbana, y ambos sexos, la distribucion de los datos es similar.

```python
conteos = datos['fecha_año'].value_counts()
plt.figure(figsize=(8,8)) 
colores = ['lightblue', 'lightpink', 'lightgreen', 'lightyellow', 'lightgray']
plt.pie(conteos, labels=conteos.index, autopct='%0.1f%%',labeldistance=1.2, colors=colores, textprops={'fontsize': 10})
plt.title('Distribución de la variable año de encuesta', y=1.1, fontsize=12) #y=1.1 cambia la posición del título

plt.axis('equal')  # Para que el gráfico sea un círculo en lugar de una elipse
plt.legend(loc='lower right', bbox_to_anchor=(1.5, 1), title='Etiquetas', fontsize=10)
plt.show()
```
